```python
import math

def rachas(data, alpha=0.05):


    r = []
    for i in range(1, 100):
        if data[i] > data[i-1]:
            r.append(1)
        else:
            r.append(0)

    h=1
    for i in range(1, len(r)):
      if r[i]!= r[i-1]:
            h += 1
    n=100
#calculamos la media
    mh=(2*n-1)/3
#calculamos varianza
    vh=(16*n-29)/90
#calculamos la desviación estándar
    dvstd=math.sqrt(vh)
#calculamos el estadístico z
    z=(h-mh)/dvstd
    print(f"Numero de elementos=",n)
    print(f"Numero de rachas observadas=",h)
    print(f"La media es=",mh)
    print(f"La varianza es=",vh)
    print(f"La desviacion estandar es=",dvstd)
    print(f"El estadístico z es=",z)

#Para alpha igual a 0.05, el Z alpha/2 es 1.96
    Z_critico=1.96
    if abs(z) > Z_critico:
        print("Rechazar H0")
    else: print("No rechazar H0")

```
